```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

```python
encoder=LabelEncoder()
y_train=encoder.fit_transform(y_train)
y_test=encoder.fit_transform(y_test)
```

```python
X_train_tensor = torch.from_numpy(X_train.astype(np.float32))
X_test_tensor = torch.from_numpy(X_test.astype(np.float32))
y_train_tensor = torch.from_numpy(y_train.astype(np.float32))
y_test_tensor = torch.from_numpy(y_test.astype(np.float32))
```

```python
X_train_tensor.shape
```

```python
y_train_tensor.shape
```

```python
from torch.utils.data import Dataset,DataLoader

class customedataset(Dataset):
    def __init__(self,features,labels):
        self.features=features
        self.lables=labels
    def __len__(self):
        return len(self.features)
    def __getitem__(self,idx):
        return self.features[idx],self.lables[idx]
```

```python
train_datasets=customedataset(X_test_tensor,y_test_tensor)
test_datasets=customedataset(X_test_tensor,y_test_tensor)
```

```python
train_datasets[10]
```

```python
train_loader=DataLoader(train_datasets,batch_size=32,shuffle=True)
test_loader=DataLoader(test_datasets,batch_size=32,shuffle=True)
```

```python
class MySimpleNN(nn.Module):

  def __init__(self, num_features):
    super().__init__()
    self.linear=nn.Linear(num_features,1)
    self.sigmoid=nn.Sigmoid()
    
  def forward(self, features):
    out=self.linear(features)
    out=self.sigmoid(out)
    return out
```

```python
learning_rate = 0.1
epochs = 100
```

```python
loss_function=nn.BCELoss()
```

```python
# create model
model = MySimpleNN(X_train_tensor.shape[1])

#define optimizer 
optimizer=torch.optim.SGD(model.parameters(),lr=learning_rate)

# define loop
for epoch in range(epochs):

  for batch_features,batch_labels in train_loader:
  # forward pass
    y_pred = model(batch_features)

    # loss calculate
    loss = loss_function(y_pred, batch_labels.view(-1,1))
    
    # zero gradients
    optimizer.zero_grad()
    
    # backward pass
    loss.backward()

    # parameters update
    optimizer.step()

    # print loss in each epoch
    print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')
```

```python
model.linear.bias
```

```python
# model evaluation
accuracy_list=[]
with torch.no_grad():
  for batch_features,batch_labels in test_loader:
    y_pred = model(batch_features)
    y_pred = (y_pred > 0.8).float()

    batch_accuracy = (y_pred.view(-1) == batch_labels).float().mean().item()
    accuracy_list.append(batch_accuracy)
  
#overall accuracy
overall_accuracy=sum(accuracy_list)/len(accuracy_list)
print(f'accuracy:{overall_accuracy:.4f}')
```
